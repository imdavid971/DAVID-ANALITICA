```python
import csv
import numpy as np

arreglo=np.array([])
with open("employees.csv") as mycsv:
    data=csv.reader(mycsv)
    for x in data:
        arreglo=np.append(arreglo,x)

arreglo=arreglo.reshape(51,11)
salario = (arreglo[1:,7:8])
salario = salario.astype(int)
print(salario)
print("-"*50)
print("La nómina completa:", np.sum(salario))
print("-"*50)
print("El Promedio Salarial es:", np.mean(salario))
print("-"*50)
print("El Salario Mas Alto es:", np.max(salario))
print("-"*50)
print("El Salario Mas Bajo es:", np.min(salario))
```
